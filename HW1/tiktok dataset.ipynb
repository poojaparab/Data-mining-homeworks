##### 1. Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

##### 2. Reading CSV file in pandas dataframe

```python
df = pd.read_csv("TikTok_songs_2019.csv")
```

##### 3. Printing first 5 rows

```python
df.head()
```

##### 4. Generating descriptive statistics

```python
df.describe(include='all')
```

##### Result observation: 15 rows are quantitative and 3 rows qualitative
    
    

##### 5. Summary of dataset

```python
print("Shape of the dataset:\n{}".format(df.shape))
print("Column names in the dataset:\n{}".format(df.columns))
print("Unique values in each column:\n{}".format(df.nunique(axis=0)))
print("Types of attributes:\n{}".format(df.info()))
```

##### 6. Checking if there is any missing value

```python
df.isna().sum()
```

##### Result Observation: There is no missing value in any column of the data

##### Alternative Method: Checking if there is any missing value

```python
df.isnull().sum()
```

##### 7. Attribute Distribution

```python
col_list = df.select_dtypes(include=['int64','float64']).columns.to_list()
plt.figure(figsize=(16, 6))
for i in col_list:
    
    df[i].plot(kind='hist', bins=25, figsize=(12,6), facecolor='green',edgecolor='black')
    plt.xlabel(i)
    plt.show()

```

##### 8. Correlation beween columns

```python
corr=df.corr()
plt.figure(figsize=(16, 6))
sns.heatmap(corr,vmin=-1, vmax=1, annot=True)
```

##### 9. Scatter plot of all the pairs

```python
plt.figure(figsize=(80,80))
sns.pairplot(df)
plt.show()
```

##### 10.Covariance between columns

```python
df.cov()
```
